# 1. Classification
Example: Suppose we have a dataset of exam scores for students and we want to predict whether a student will pass or fail the exam. The target variable here is discrete because it can only take two values: "Pass" or "Fail."

Binary Classification Example:
Input Feature: Exam Score (out of 100)
Target Variable: Pass (1) or Fail (0)

```python
import pandas as pd

# 1. Binary Classification Example

# Creating a DataFrame for Binary Classification
binary_classification_data = {
    "Exam Score": [85, 40, 75, 55, 92],
    "Pass/Fail (Target)": [1, 0, 1, 0, 1]
}
binary_classification_df = pd.DataFrame(binary_classification_data)

print("Binary Classification Example:")
print(binary_classification_df)
print("\n")
```

## Graphical Representation:

In binary classification, you can imagine a threshold on the exam score (e.g., 50), where scores above 50 result in a "Pass" (1) and scores below 50 result in a "Fail" (0).

```python
import matplotlib.pyplot as plt

# Data points
scores = [85, 40, 75, 55, 92]
results = [1, 0, 1, 0, 1]

# Plotting
plt.scatter(scores, results, color='blue')
plt.axhline(y=0.5, color='red', linestyle='--', label='Threshold')
plt.xlabel('Exam Score')
plt.ylabel('Pass/Fail')
plt.title('Binary Classification')
plt.yticks([0, 1], ['Fail', 'Pass'])
plt.legend()
plt.show()
```

# 2. Multi-class Classification
Example: Now, let's consider a dataset where we want to classify the type of fruit based on its features like weight and color. The target variable is discrete but with more than two possible outcomes, such as "Apple," "Banana," and "Cherry."

Multi-class Classification Example:
Input Features: Weight, Color
Target Variable: Fruit Type (Apple, Banana, Cherry)

```python
# 2. Multi-class Classification Example

# Creating a DataFrame for Multi-class Classification
multi_class_classification_data = {
    "Weight (g)": [150, 120, 10, 200, 100],
    "Color (RGB)": ["(255, 0, 0)", "(255, 255, 0)", "(255, 0, 0)", "(255, 0, 0)", "(255, 255, 0)"],
    "Fruit Type (Target)": ["Apple", "Banana", "Cherry", "Apple", "Banana"]
}
multi_class_classification_df = pd.DataFrame(multi_class_classification_data)

print("Multi-class Classification Example:")
print(multi_class_classification_df)
print("\n")

# 3. Regression Example

# Creating a DataFrame for Regression
regression_data = {
    "House Size (sq ft)": [1500, 2000, 2500, 3000, 3500],
    "House Price ($)": [300000, 400000, 500000, 600000, 700000]
}
regression_df = pd.DataFrame(regression_data)

print("Regression Example:")
print(regression_df)
```

## Graphical Representation:

In a regression task, the goal is to fit a line (or curve) through the data points that best predicts the target variable (house price) based on the input feature (house size).
Here's a simple plot to visualize this:

```python
import numpy as np

# Data points
sizes = np.array([1500, 2000, 2500, 3000, 3500])
prices = np.array([300000, 400000, 500000, 600000, 700000])

# Fitting a line
fit = np.polyfit(sizes, prices, 1)
fit_fn = np.poly1d(fit)

# Plotting
plt.scatter(sizes, prices, color='green')
plt.plot(sizes, fit_fn(sizes), color='blue', linestyle='--')
plt.xlabel('House Size (sq ft)')
plt.ylabel('House Price ($)')
plt.title('Regression Example')
plt.show()
```
